**Pandas Built-in Data Visualization**

```python
import numpy as np
import pandas as pd
import matplotlib 
```

```python
df1 = pd.read_csv('df1',index_col=0)
df2 = pd.read_csv('df2')
```

#### Style Sheets ####

* Before the plot look like this

```python
df1.head()
```

```python
df1['A'].hist()
```

* Now we call the style

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

* And finally the plot looks like this

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

#### Area ####

```python
df2.plot.area(alpha=0.4)
```

#### Barplots ####

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

#### Histograms ####

```python
df1['A'].plot.hist(bins=50)
```

#### Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x='A',y='B',s=df1['C']*200)
```

#### BoxPlots

```python
df2.plot.box() # Can also pass a by= argument for groupby
```
